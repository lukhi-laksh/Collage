```python
import pandas as pd 
import seaborn as sns  
data = pd.read_csv('Employee_dataset.csv') 
print(data.head()) 
```

```python
x = data['STATUS_YEAR'].value_counts() 
plt.pie(x.values, 
        labels=x.index, 
        autopct='%1.1f%%') 
plt.show() 
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(15, 5)) 
sns.barplot(x=data['department_name'], y=data['length_of_service']) 
plt.xticks(rotation=90) 
```

```python
sns.scatterplot(x=data['length_of_service'], 

                y=data['age']) 
```

```python
import seaborn as sns
sns.countplot(x='STATUS_YEAR', hue='STATUS', data=data)
```

```python
from sklearn import datasets, decomposition 
iris = datasets.load_iris() 
X = iris.data 
y = iris.target 
pca = decomposition.PCA(n_components=2) 
X = pca.fit_transform(X) 
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y) 
```
