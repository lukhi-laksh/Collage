```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample movie data
data = {
    "movieId": [1, 2, 3, 4, 5],
    "title": [
        "Toy Story (1995)",
        "Jumanji (1995)",
        "Grumpier Old Men (1995)",
        "Waiting to Exhale (1995)",
        "Father of the Bride Part II (1995)"
    ],
    "genres": [
        "Adventure|Animation|Children|Comedy|Fantasy",
        "Adventure|Children|Fantasy",
        "Comedy|Romance",
        "Comedy|Drama|Romance",
        "Comedy"
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Process genres
genre_counts = df['genres'].str.split('|').explode().value_counts()

# Set up plotting area for multiple plots
fig, axs = plt.subplots(3, 2, figsize=(16, 16))
fig.suptitle('Different Types of Genre Distribution Graphs', fontsize=16)

# 1. Bar Chart
axs[0, 0].bar(genre_counts.index, genre_counts.values, color='skyblue')
axs[0, 0].set_title('Bar Chart')
axs[0, 0].set_ylabel('Count')
axs[0, 0].tick_params(axis='x', rotation=45)

# 2. Pie Chart
axs[0, 1].pie(genre_counts.values, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
axs[0, 1].set_title('Pie Chart')

# 3. Line Plot
axs[1, 0].plot(genre_counts.index, genre_counts.values, marker='o', color='green')
axs[1, 0].set_title('Line Plot')
axs[1, 0].set_ylabel('Count')
axs[1, 0].tick_params(axis='x', rotation=45)

# 4. Horizontal Bar Chart
axs[1, 1].barh(genre_counts.index, genre_counts.values, color='salmon')
axs[1, 1].set_title('Horizontal Bar Chart')
axs[1, 1].set_xlabel('Count')

# 5. Stacked Bar Chart (Genres per movie)
genres_df = df['genres'].str.get_dummies(sep='|')
genres_df.index = df['title']
genres_df.plot(kind='bar', stacked=True, colormap='tab20', ax=axs[2, 0])
axs[2, 0].set_title('Stacked Bar Chart (Genres per Movie)')
axs[2, 0].set_ylabel('Presence')
axs[2, 0].tick_params(axis='x', rotation=45)

# 6. Area Chart
axs[2, 1].fill_between(range(len(genre_counts)), genre_counts.values, color='lightgreen', alpha=0.6)
axs[2, 1].set_xticks(range(len(genre_counts)))
axs[2, 1].set_xticklabels(genre_counts.index, rotation=45)
axs[2, 1].set_title('Area Chart')
axs[2, 1].set_ylabel('Count')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "movieId": [1, 2, 3, 4, 5],
    "title": [
        "Toy Story (1995)",
        "Jumanji (1995)",
        "Grumpier Old Men (1995)",
        "Waiting to Exhale (1995)",
        "Father of the Bride Part II (1995)"
    ],
    "genres": [
        "Adventure|Animation|Children|Comedy|Fantasy",
        "Adventure|Children|Fantasy",
        "Comedy|Romance",
        "Comedy|Drama|Romance",
        "Comedy"
    ]
}

df = pd.DataFrame(data)
genre_counts = df['genres'].str.split('|').explode().value_counts()

plt.figure(figsize=(8,5))
plt.bar(genre_counts.index, genre_counts.values, color='skyblue')
plt.title('Genre Distribution - Bar Chart')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(genre_counts.values, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Genre Distribution - Pie Chart')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(genre_counts.index, genre_counts.values, marker='o', color='green')
plt.title('Genre Distribution - Line Plot')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.barh(genre_counts.index, genre_counts.values, color='salmon')
plt.title('Genre Distribution - Horizontal Bar Chart')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

```python
genres_df = df['genres'].str.get_dummies(sep='|')
genres_df.index = df['title']

genres_df.plot(kind='bar', stacked=True, figsize=(10,6), colormap='tab20')
plt.title('Genre Presence Per Movie - Stacked Bar Chart')
plt.xlabel('Movie Title')
plt.ylabel('Presence (1 = Yes)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.fill_between(range(len(genre_counts)), genre_counts.values, color='lightgreen', alpha=0.6)
plt.xticks(range(len(genre_counts)), genre_counts.index, rotation=45)
plt.title('Genre Distribution - Area Chart')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()
```
