# Q1. Perform data cleaning operations according to following.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## A. Import data and identify missing values and create visualization of it. 

```python
df = pd.read_csv("netflix_titles.csv")
x = df.columns
y = df.isnull().sum()
plt.bar(x, y)
plt.xticks(rotation=90)
plt.ylabel('Number of Null values') 

plt.xlabel('Fields') 

plt.title("Showing Null values count") 

plt.grid(False)
plt.show()
```

## B. drop duplicate values from dataset.

```python
df.drop_duplicates()
print("after Droped:-")
df
```

## C. Remove the NaN/Null values from the dataset. 

```python
df.dropna()
print("After Drop Null :-")
df
```

## D. Transform column values of rating according to following.  
    'TV-PG': 'Older Kids',  
    'TV-MA': 'Adults',  
    'TV-Y7-FV': 'Older Kids',  
    'TV-Y7': 'Older Kids', 
    'TV-14': 'Teens',  
    'R': 'Adults',  
    'TV-Y': 'Kids',  
    'NR': 'Adults',  
    'PG-13': 'Teens',  
    'TV-G': 'Kids',  
    'PG': 'Older Kids',  
    'G': 'Kids',  
    'UR': 'Adults',  
    'NC-17': 'Adults' 

```python
rating_map = {
    'TV-PG': 'Older Kids',
    'TV-MA': 'Adults',
    'TV-Y7-FV': 'Older Kids',
    'TV-Y7': 'Older Kids',
    'TV-14': 'Teens',
    'R': 'Adults',
    'TV-Y': 'Kids',
    'NR': 'Adults',
    'PG-13': 'Teens',
    'TV-G': 'Kids',
    'PG': 'Older Kids',
    'G': 'Kids',
    'UR': 'Adults',
    'NC-17': 'Adults'
}

df['rating'] = df['rating'].replace(rating_map)
df
```

# Q2. Analyse the dataset and answer following questions by performing exploratory data analysis and data visualization. 

## A. Identify the content distribution of TV and movie shows and plot graph.  

```python
df['type'].value_counts().plot(kind="line")
plt.show()
```

## B. Find out the COUNTRIES WITH HIGHEST NUMBER OF MOVIES & TV-SHOWS.

```python
df['country'].value_counts().head(10) 
```

## C. Identify the percentage of CONTENT ADDED OVERS YEARS.

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
content_per_year = df['date_added'].dt.year.value_counts(normalize=True) * 100
content_per_year.sort_index() 
```

## D. Find out the director and actors with highest shows. 

```python
director_counts = df['director'].value_counts()
top_director = director_counts.idxmax()
top_director_count = director_counts.max()
print(f"Director with most shows: {top_director} ({top_director_count} shows)")
```

## E. What kind of a shows are more popular in a country. Plot the graph of genre wise popular shows. 

```python
country = 'United States'
country_df = df[df['country'] == country]
genre_df = country_df['listed_in'].str.split(',', expand=True).stack().reset_index(drop=True)
genre_df = genre_df.str.strip()
genre_counts = genre_df.value_counts()
print(genre_counts)


plt.figure(figsize=(12,6))
genre_counts.plot(kind='bar', color='orange')
plt.title(f"Genre-wise Popular Shows in {country}")
plt.xlabel("Genre")
plt.ylabel("Number of Shows")
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

## F. Identify top 5 actors of movies and TV shows based on the number of titles. 

```python
df_cast = df.dropna(subset=['cast'])
df_cast = df_cast.assign(actor=df_cast['cast'].str.split(',')).explode('actor')
df_cast['actor'] = df_cast['actor'].str.strip()

top_actors_movies = df_cast[df_cast['type']=='Movie']['actor'].value_counts().head(5)
print("Top 5 Actors in Movies:")
print(top_actors_movies)


top_actors_tv = df_cast[df_cast['type']=='TV Show']['actor'].value_counts().head(5)
print("\nTop 5 Actors in TV Shows:")
print(top_actors_tv)
```
